```python
import pickle
import numpy as np

with open(f'data/realworld.pkl', 'rb') as f:
    x_rw, y_rw, k_rw = pickle.load(f)
print('Realworld')
print(x_rw.shape, y_rw.shape, k_rw.shape)
print(np.unique(y_rw), np.unique(k_rw))

with open(f'data/mobiact.pkl', 'rb') as f:
    x_ma, y_ma, k_ma = pickle.load(f)
print('\nMobiAct')
print(x_ma.shape, y_ma.shape, k_ma.shape)
print(np.unique(y_ma), np.unique(k_ma))

with open(f'data/pamap.pkl', 'rb') as f:
    x_pa, y_pa, k_pa = pickle.load(f)
print('\nPAMAP')
print(x_pa.shape, y_pa.shape, k_pa.shape)
print(np.unique(y_pa), np.unique(k_pa))
```

```python
import matplotlib.pyplot as plt
import random

random.seed(2710)

classes = ['Walking', 'Running', 'Climbing Down', 'Climbing Up']

def plot_dataset(x, y, k, title):

    # Initialize the figure
    n_subjects = 4
    n_activities = 4
    fig, axes = plt.subplots(n_subjects, n_activities, figsize=(n_activities*2, n_subjects), sharex=True, sharey=True)

    # Loop over each subject
    for subject in range(n_subjects):
        # Loop over each activity
        for activity in range(n_activities):
            # Get the indices for the current subject and activity
            indices = np.where((k == subject+1) & (y == activity))[0]
            
            # If there are any matching indices, select one at random
            if len(indices) > 0:
                idx = random.choice(indices)
                time_series = x[idx]
                
                # Plot the time series for each channel
                for channel in range(3):
                    axes[subject, activity].plot(time_series[channel], label=f'Channel {channel+1}', linewidth=0.75)
            
            # Remove the x and y ticks for clarity
            axes[subject, activity].get_xaxis().set_visible(False)
            axes[subject, activity].get_yaxis().set_visible(False)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.savefig(f'plots/{title}.png', dpi=100)
    plt.show()

plot_dataset(x_rw, y_rw, k_rw, 'rw01')
plot_dataset(x_ma, y_ma, k_ma, 'ma01')
plot_dataset(x_pa, y_pa, k_pa, 'pa01')
```
