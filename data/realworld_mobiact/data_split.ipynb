```python
from sklearn.model_selection import train_test_split
import torch
import pickle

with open('realworld_mobiact_df.pkl', 'rb') as f:
    x_df, y_df, k_df = pickle.load(f)

print(x_df.shape, y_df.shape, k_df.shape)

x_df_train, x_df_test, y_df_train, y_df_test = train_test_split(x_df, y_df, test_size=0.2, random_state=2710, stratify=y_df, shuffle=True)
print(x_df_train.shape, x_df_test.shape, y_df_train.shape, y_df_test.shape)

train_df = {'samples': x_df_train, 'labels': y_df_train}
test_df = {'samples': x_df_test, 'labels': y_df_test}

torch.save(train_df, 'train_df.pt')
torch.save(test_df, 'test_df.pt')
```

```python
import numpy as np

with open('realworld_mobiact_dp.pkl', 'rb') as f:
    x_dp, y_dp, k_dp = pickle.load(f)

print(x_dp.shape, y_dp.shape, k_dp.shape)

for domain in np.unique(k_dp):
    mask = k_dp == domain
    x_dp_dom = x_dp[mask]
    y_dp_dom = y_dp[mask]

    x_dp_train, x_dp_test, y_dp_train, y_dp_test = train_test_split(x_dp_dom, y_dp_dom, test_size=0.2, random_state=2710, stratify=y_dp_dom, shuffle=True)
    print(x_dp_train.shape, x_dp_test.shape, y_dp_train.shape, y_dp_test.shape)

    train_dp = {'samples': x_dp_train, 'labels': y_dp_train}
    test_dp = {'samples': x_dp_test, 'labels': y_dp_test}

    torch.save(train_dp, f'train_dp_{domain}.pt')
    torch.save(test_dp, f'test_dp_{domain}.pt')
```
